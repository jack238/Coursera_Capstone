```python
import pandas as pd
df = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")[0]
df.head()
```

#import pandas and read table into df 

```python
indexNames = df[ df['Borough'] == "Not assigned" ].index
df.drop(indexNames , inplace=True)
df.head()
```

#Remove all row where Borough = Not assigned

#replace where Neighbourhood = Not assigned with Borough and sort df by Postcode

```python
df['Neighbourhood'].loc[df['Neighbourhood'] == "Not assigned"] = df['Borough']
#add sort dataframe by postcode column
df.sort_values(by=['Postcode'])
```

#use groupby and join to merge duplicated Postcode rows and print shape

```python
df = df.groupby(['Postcode','Borough'])['Neighbourhood'].apply(', '.join).reset_index()
shape = df.shape
print (shape)
```
